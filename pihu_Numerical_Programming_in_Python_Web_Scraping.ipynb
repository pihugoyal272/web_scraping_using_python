````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
def fetch_movie_urls(url):
  response = requests.get(url, headers=headers)
  if response.status_code != 200:
    return "Failed to retrieve the page, status code:", response.status_code
  soup = BeautifulSoup(response.text, 'html.parser')
  return soup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}


url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
soup=fetch_movie_urls(url)
print(soup.prettify())

## Hint : Use the following code to extract the film urls
# movie_links = soup.find_all('a', href=True)
# movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

# url_list=[]
# for x in movie_urls:
#   url_list.append('https://www.justwatch.com'+x)
````

## **Fetching Movie URL's**

````python
# Write Your Code here
movie_url = []
for i in soup.find_all('a', class_ = 'title-list-grid__item--link'):
  movie_url.append('https://www.justwatch.com' +i['href'])
print(movie_url)
print(f'Total number of Movie is {len(movie_url)}')
````

## **Scrapping Movie Title**

````python
movie_title = []
for url in movie_url:
  try:
    content = requests.get(url,headers = headers)
    soup = BeautifulSoup(content.text,'html.parser')
    title = soup.find_all('h1')[0].text
  except:
    title = 'NA'
  movie_title.append(title)
print(movie_title)
print('Total number of title is' , len(movie_title))
````

## **Scrapping release Year**

````python
# Write Your Code here
movie_year = []
for url in movie_url:
  try:
    content = requests.get(url,headers = headers)
    soup = BeautifulSoup(content.text, 'html.parser')
    year = soup.find_all('span' , class_= 'release-year')[0].text.strip('()')
  except:
    year = 'NA'
  movie_year.append(year)
print(movie_year)
print('Total number of movie:', len(movie_year))
````

## **Scrapping Genres**

````python
movie_genres = []
for url in movie_url:
  try:
    content = requests.get(url,headers = headers)
    soup = BeautifulSoup(content.text,'html.parser')
    for i in soup.find_all('div', class_ = 'detail-infos'):
      if i.find_all('h3')[0].text == 'Genres':
        genres = i.find_all('span')[0].text
  except:
    genres = 'NA'
  movie_genres.append(genres)
print(movie_genres)
print(f'Total number of genres is : {len(movie_genres)}')
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
imdb_rating = []
for url in movie_url:
  try:
    content = requests.get(url, headers = headers)
    soup = BeautifulSoup(content.text,'html.parser')
    for i in soup.find_all('div', class_ = 'detail-infos'):
      if i.find_all('h3')[0].text =='Rating':
        rating = i.find_all('div')[0].text.strip().split()[0]
  except:
    rating = 'NA'
  imdb_rating.append(rating)
print(imdb_rating)
print(f'Number of imdb_rating is : {len(imdb_rating)}')
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
movie_runtime = []
for url in movie_url:
  try:
    content = requests.get(url,headers = headers)
    soup = BeautifulSoup(content.text,'html.parser')
    for i in soup.find_all('div',class_ = 'detail-infos'):
      if i.find_all('h3')[0].text == 'Runtime':
        runtime = i.find_all('div')[0].text
  except:
    runtime = 'NA'
  movie_runtime.append(runtime)
print(movie_runtime)
print('The number of runtime :' , len(movie_runtime))
````

## **Scrapping Age Rating**

````python
# Write Your Code here
movie_age_rating = []
for url in movie_url:
  content = requests.get(url,headers = headers)
  soup = BeautifulSoup(content.text,'html.parser')
  age_rating = 'NA'
  for i in soup.find_all('div', class_ = 'detail-infos'):
    if i.find_all('h3')[0].text == 'Age rating' :
      age_rating = i.find_all('div')[0].text
      break
  movie_age_rating.append(age_rating)
print(movie_age_rating)
print('The number of rating is :', len(movie_age_rating))
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
movie_production_country = []
for url in movie_url:
  try:
    content = requests.get(url,headers = headers)
    soup = BeautifulSoup(content.text,'html.parser')
    for i in soup.find_all('div' , class_ = 'detail-infos'):
      if i.find_all('h3')[0].text == ' Production country ':
        country = i.find_all('div')[0].text
  except:
    country = 'NA'
  movie_production_country.append(country)
print(movie_production_country)
print('The number of movie production country is:' , len(movie_production_country))
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
movie_stream_service=[]
for url in movie_url:
  try:
     content=requests.get(url,headers=headers)
     soup=BeautifulSoup(content.text,'html.parser')
     names = [i['alt'] for i in soup.find_all('img', class_ = 'offer__icon')]
  except:
    names = 'NA'
  movie_stream_service.append(','.join(names))
print(movie_stream_service)
````

## **Now Creating Movies DataFrame**

````python
# Write Your Code here

info = {'Movie_URL' : movie_url, 'Movie_Name' : movie_title, 'Release_Year' : movie_year, 'Genres' : movie_genres, 'Runtime' : movie_runtime,
        'Age_Rating' : movie_age_rating, 'Production_Country' : movie_production_country, 'Rating' : imdb_rating , 'Stream_Service' : movie_stream_service }
data_movie = pd.DataFrame(info)
data_movie
````
